```python
%%sql

SELECT 
  TO_CHAR(st.ts, 'YYYY-MM') as month, 
  usc.channel,
  COUNT(DISTINCT usc.userid) as cnt
FROM
  raw_data.session_timestamp st
  JOIN raw_data.user_session_channel usc ON st.sessionid = usc.sessionid
GROUP BY 1, 2
ORDER BY 1 DESC, 2
```

```python
%%sql

CREATE TABLE plerin152.monthly_active_user_summary AS
(
  SELECT TO_CHAR(ts, 'YYYY-MM') AS month, COUNT(DISTINCT userid)
  FROM raw_data.user_session_channel A
  JOIN raw_data.session_timestamp B ON A.sessionid = B.sessionid
  GROUP BY 1
  ORDER BY 1 DESC
);
```
